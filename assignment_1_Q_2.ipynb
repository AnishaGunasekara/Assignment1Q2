```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import argparse

from __future__ import print_function

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/pic_2.jpg', cv.IMREAD_COLOR)
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2RGB)

def adjust_gamma (im_og, gamma=0.5):

    table = np.array([((i/255.0)**gamma)*255.0
                   for i in np.arange(0,256)]).astype('uint8')

    im_gm = cv.LUT(im_og, table)
    im_gm = cv.cvtColor(im_gm, cv.COLOR_BGR2RGB)

    return im_gm

im_gm = adjust_gamma(im_og)


fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(im_og)
ax[0].set_title('Before Gamma Correction')

ax[1].imshow(im_gm)
ax[1].set_title('After Gamma Correction')

color = ('b', 'g', 'r')


fig, ax_hist = plt.subplots()
for i, c in enumerate(color):
    ax_hist.plot(cv.calcHist([im_og], [i], None, [256], [0, 256]), color=c)
    ax_hist.set_title('Before Gamma Correction')

fig, ax_hist = plt.subplots()

for i, c in enumerate(color):
    ax_hist.plot(cv.calcHist([im_gm], [i], None, [256], [0, 256]), color=c)
    ax_hist.set_title('After Gamma Correction')

plt.show()
```
